```python
import os
import pandas as pd
import numpy as np
import pickle
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms
from PIL import Image
from transformers import (
    BartTokenizer,
    BartForConditionalGeneration,
    ViTModel
)

import nltk
nltk.download("punkt")
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
import evaluate
from transformers.modeling_outputs import BaseModelOutput
import random
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

DATA_DIR    = "/kaggle/input/sarcasm/MORE-PLUS-DATASET"
TRAIN_TSV   = os.path.join(DATA_DIR, "/content/MORE-PLUS-DATASET/MORE-PLUS-DATASET/train_df.tsv")
VAL_TSV     = os.path.join(DATA_DIR, "/content/MORE-PLUS-DATASET/MORE-PLUS-DATASET/val_df.tsv")
D_TRAIN_PKL = os.path.join(DATA_DIR, "/content/MORE-PLUS-DATASET/MORE-PLUS-DATASET/D_train.pkl")
O_TRAIN_PKL = os.path.join(DATA_DIR, "/content/MORE-PLUS-DATASET/MORE-PLUS-DATASET/O_train.pkl")
D_VAL_PKL   = os.path.join(DATA_DIR, "/content/MORE-PLUS-DATASET/MORE-PLUS-DATASET/D_val.pkl")
O_VAL_PKL   = os.path.join(DATA_DIR, "/content/MORE-PLUS-DATASET/MORE-PLUS-DATASET/O_val.pkl")
IMAGES_DIR  = os.path.join(DATA_DIR, "/content/MORE-PLUS-DATASET/MORE-PLUS-DATASET/tsvimages")

train_df = pd.read_csv(TRAIN_TSV, sep="\t")
val_df   = pd.read_csv(VAL_TSV,   sep="\t")

print("Train columns:", train_df.columns)

with open(D_TRAIN_PKL, "rb") as f:
    D_train = pickle.load(f)
with open(O_TRAIN_PKL, "rb") as f:
    O_train = pickle.load(f)

with open(D_VAL_PKL, "rb") as f:
    D_val = pickle.load(f)
with open(O_VAL_PKL, "rb") as f:
    O_val = pickle.load(f)

image_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

tokenizer = BartTokenizer.from_pretrained("facebook/bart-base")

class MorePlusDataset(Dataset):
    def __init__(self, df, D_dict, O_dict, tokenizer, img_dir, transform, max_length=128):
        self.df = df.reset_index(drop=True)
        self.D_dict = D_dict
        self.O_dict = O_dict
        self.tokenizer = tokenizer
        self.img_dir = img_dir
        self.transform = transform
        self.max_length = max_length

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        pid         = row["pid"]
        caption     = str(row["text"])
        target      = str(row["target_of_sarcasm"])
        explanation = str(row["explanation"])

        desc = self.D_dict.get(pid, "")
        objects_list = self.O_dict.get(pid, [])
        objects_str  = " ".join(objects_list)

        text_input = caption + " " + desc + " " + objects_str + " </s> " + target

        enc = self.tokenizer(
            text_input,
            padding="max_length",
            truncation=True,
            max_length=self.max_length,
            return_tensors="pt"
        )
        input_ids      = enc["input_ids"].squeeze(0)
        attention_mask = enc["attention_mask"].squeeze(0)

        dec = self.tokenizer(
            explanation,
            padding="max_length",
            truncation=True,
            max_length=self.max_length // 2,
            return_tensors="pt"
        )
        labels = dec["input_ids"].squeeze(0)
        labels[labels == self.tokenizer.pad_token_id] = -100

        img_path_jpg = os.path.join(self.img_dir, f"{pid}.jpg")
        img_path_png = os.path.join(self.img_dir, f"{pid}.png")
        if os.path.exists(img_path_jpg):
            img_path = img_path_jpg
        elif os.path.exists(img_path_png):
            img_path = img_path_png
        else:
            img_path = None

        if img_path and os.path.exists(img_path):
            image = Image.open(img_path).convert("RGB")
        else:
            image = Image.fromarray(np.uint8(np.random.rand(224,224,3)*255))

        if self.transform:
            image = self.transform(image)

        return {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
            "labels": labels,
            "pid": pid,
            "image": image,
            "gold_explanation": explanation
        }

train_dataset = MorePlusDataset(train_df, D_train, O_train, tokenizer, IMAGES_DIR, image_transform)
val_dataset   = MorePlusDataset(val_df,   D_val,   O_val,   tokenizer, IMAGES_DIR, image_transform)

train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=4, shuffle=False)


class SharedFusion(nn.Module):

    def __init__(self, d_model):
        super().__init__()
        self.text_attn  = nn.MultiheadAttention(embed_dim=d_model, num_heads=4, batch_first=True)
        self.image_attn = nn.MultiheadAttention(embed_dim=d_model, num_heads=4, batch_first=True)
        self.W_t = nn.Linear(d_model, d_model)
        self.W_v = nn.Linear(d_model, d_model)
        self.sigmoid = nn.Sigmoid()

    def forward(self, E_t, E_v):
        A_t, _ = self.text_attn(E_t, E_t, E_t)
        A_v, _ = self.image_attn(E_v, E_v, E_v)
        B, seq_len, d_model = E_t.shape
        A_v_pooled   = A_v.mean(dim=1, keepdim=True)
        A_v_expanded = A_v_pooled.expand(B, seq_len, d_model)
        G_t = self.sigmoid(self.W_t(E_t.mean(dim=1)))
        G_v = self.sigmoid(self.W_v(A_v_pooled.squeeze(1)))
        G_t = G_t.unsqueeze(1)
        G_v = G_v.unsqueeze(1)
        fused = A_t + (G_v * A_v_expanded)
        return fused


class TurboNoKGNoGCN(nn.Module):
    """
    TURBO-like Model: BART (base) + ViT + Shared Fusion.
    """
    def __init__(self, bart_name="facebook/bart-base", vit_name="google/vit-base-patch16-224", d_model=768):
        super().__init__()
        self.bart = BartForConditionalGeneration.from_pretrained(bart_name)
        self.bart.config.forced_bos_token_id = None

        self.vit = ViTModel.from_pretrained(vit_name)
        self.img_proj = nn.Linear(self.vit.config.hidden_size, d_model)

        self.shared_fusion = SharedFusion(d_model)
        self.d_model = d_model

    def forward(self, input_ids, attention_mask, images, labels=None):
        encoder_outputs = self.bart.model.encoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            return_dict=True
        )
        text_embeds = encoder_outputs.last_hidden_state

        vit_outputs = self.vit(images)
        image_embeds = vit_outputs.last_hidden_state
        image_embeds = self.img_proj(image_embeds)

        fused_enc = self.shared_fusion(text_embeds, image_embeds)

        outputs = self.bart(
            encoder_outputs=(fused_enc,),
            attention_mask=attention_mask,
            labels=labels,
            return_dict=True
        )
        return outputs

    def generate_explanation(self, input_ids, attention_mask, images, max_length=50, num_beams=2):
        encoder_outputs = self.bart.model.encoder(
            input_ids=input_ids,
            attention_mask=attention_mask,
            return_dict=True
        )
        text_embeds = encoder_outputs.last_hidden_state

        vit_outputs = self.vit(images)
        image_embeds = vit_outputs.last_hidden_state
        image_embeds = self.img_proj(image_embeds)

        fused_enc = self.shared_fusion(text_embeds, image_embeds)
        fused_encoder_outputs = BaseModelOutput(last_hidden_state=fused_enc)

        gen_ids = self.bart.generate(
            encoder_outputs=fused_encoder_outputs,
            attention_mask=attention_mask,
            max_length=max_length,
            num_beams=num_beams,
            early_stopping=True,
            no_repeat_ngram_size=2
        )
        return gen_ids

model = TurboNoKGNoGCN().to(device)
optimizer = optim.AdamW(model.parameters(), lr=1e-4)
num_epochs = 10
save_dir   = "./checkpoints"
os.makedirs(save_dir, exist_ok=True)

def train_one_epoch(model, dataloader, optimizer, device):
    model.train()
    total_loss = 0
    for batch in dataloader:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        images = batch["image"].to(device)
        labels = batch["labels"].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask, images, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(dataloader)

def eval_one_epoch(model, dataloader, device):
    model.eval()
    total_loss = 0
    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            images = batch["image"].to(device)
            labels = batch["labels"].to(device)
            outputs = model(input_ids, attention_mask, images, labels=labels)
            loss = outputs.loss
            total_loss += loss.item()
    return total_loss / len(dataloader)

def compute_metrics(model, dataloader, device, tokenizer):
    model.eval()
    all_predictions = []
    all_references = []
    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            images = batch["image"].to(device)
            gen_ids = model.generate_explanation(input_ids, attention_mask, images, max_length=50, num_beams=2)
            gen_texts = [tokenizer.decode(g, skip_special_tokens=True) for g in gen_ids]
            gold_texts = batch["gold_explanation"]
            all_predictions.extend(gen_texts)
            all_references.extend(gold_texts)

    # Compute ROUGE
    rouge_metric = evaluate.load("rouge")
    results_rouge = rouge_metric.compute(predictions=all_predictions, references=all_references)

    # Compute BLEU (using NLTK)
    smooth_fn = SmoothingFunction().method1
    bleu1_scores = []
    bleu2_scores = []
    bleu3_scores = []
    bleu4_scores = []
    for ref, hyp in zip(all_references, all_predictions):
        ref_tokens = nltk.word_tokenize(ref)
        hyp_tokens = nltk.word_tokenize(hyp)
        bleu1 = sentence_bleu([ref_tokens], hyp_tokens, weights=(1,0,0,0), smoothing_function=smooth_fn)
        bleu1_scores.append(bleu1)
        bleu2 = sentence_bleu([ref_tokens], hyp_tokens, weights=(0.5,0.5,0,0), smoothing_function=smooth_fn)
        bleu2_scores.append(bleu2)
        bleu3 = sentence_bleu([ref_tokens], hyp_tokens, weights=(0.33,0.33,0.34,0), smoothing_function=smooth_fn)
        bleu3_scores.append(bleu3)
        bleu4 = sentence_bleu([ref_tokens], hyp_tokens, weights=(0.25,0.25,0.25,0.25), smoothing_function=smooth_fn)
        bleu4_scores.append(bleu4)

    # Compute METEOR
    meteor_metric = evaluate.load("meteor")
    results_meteor = meteor_metric.compute(predictions=all_predictions, references=all_references)

    # Compute BERTScore
    bert_scorer = evaluate.load("bertscore")
    results_bert = bert_scorer.compute(predictions=all_predictions,
                                       references=all_references,
                                       model_type="bert-base-uncased")
    metrics = {
        "rouge": results_rouge,
        "bleu1": np.mean(bleu1_scores),
        "bleu2": np.mean(bleu2_scores),
        "bleu3": np.mean(bleu3_scores),
        "bleu4": np.mean(bleu4_scores),
        "meteor": results_meteor["meteor"],
        "bert_precision": np.mean(results_bert["precision"]),
        "bert_recall": np.mean(results_bert["recall"]),
        "bert_f1": np.mean(results_bert["f1"]),
        "predictions": all_predictions,
        "references": all_references
    }
    return metrics

best_val_loss = float("inf")
best_epoch = -1

for epoch in range(num_epochs):
    train_loss = train_one_epoch(model, train_loader, optimizer, device)
    val_loss   = eval_one_epoch(model, val_loader, device)
    print(f"Epoch {epoch+1}/{num_epochs} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")

    metrics = compute_metrics(model, val_loader, device, tokenizer)
    print("\n--- Evaluation Metrics after Epoch {} ---".format(epoch+1))
    print(f"ROUGE-1: {metrics['rouge']['rouge1']:.4f}")
    print(f"ROUGE-2: {metrics['rouge']['rouge2']:.4f}")
    print(f"ROUGE-L: {metrics['rouge']['rougeL']:.4f}")
    print(f"BLEU-1: {metrics['bleu1']:.4f}")
    print(f"BLEU-2: {metrics['bleu2']:.4f}")
    print(f"BLEU-3: {metrics['bleu3']:.4f}")
    print(f"BLEU-4: {metrics['bleu4']:.4f}")
    print(f"METEOR: {metrics['meteor']:.4f}")
    print(f"BERT Precision: {metrics['bert_precision']:.4f}")
    print(f"BERT Recall: {metrics['bert_recall']:.4f}")
    print(f"BERT F1: {metrics['bert_f1']:.4f}")

    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_epoch = epoch + 1
        best_checkpoint_path = os.path.join(save_dir, "best_checkpoint.pt")
        torch.save({
            "epoch": epoch+1,
            "model_state_dict": model.state_dict(),
            "optimizer_state_dict": optimizer.state_dict()
        }, best_checkpoint_path, _use_new_zipfile_serialization=False)
        print(f"Best model updated at epoch {epoch+1} with Val Loss: {val_loss:.4f}")

print(f"\nTraining complete. Best model found at epoch {best_epoch} with Val Loss: {best_val_loss:.4f}")


```
